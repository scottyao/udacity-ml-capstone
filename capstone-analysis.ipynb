## Data Exploration

```python
import pandas as pd
import numpy as np
from IPython.display import display
import matplotlib.pyplot as plt

%matplotlib inline

train_data = pd.read_csv('train.csv')
test_data = pd.read_csv('test.csv')

display(train_data.head())

display(train_data.describe())

display(test_data.shape)
```

## Data Cleasing

```python
# remove constant columns (std = 0)
remove = []
for col in train_data.columns:
    if train_data[col].std() == 0:
        remove.append(col)
print "Dropped %d columns: %s" % (len(remove), str(remove))

# drop constant columns
train_data.drop(remove, axis=1, inplace=True)
test_data.drop(remove, axis=1, inplace=True)
```

```python
# clean var3
print train_data['var3'].value_counts()[0:10]

# replace -99999 with 2 to remove outliers
train_data.loc[train_data.var3 == -999999, 'var3'] = 2
print train_data['var3'].value_counts()[0:10]
```

```python
# remove duplicate columns
duplicate_col = []
cols = train_data.columns
for i in range(len(cols)-1):
    v = train_data[cols[i]].values
    for j in range(i+1,len(cols)):
        if np.array_equal(v,train_data[cols[j]].values):
            duplicate_col.append(cols[j])
print "Train data set has %d duplicate columns %s" % (len(duplicate_col), str(duplicate_col))

# drop duplicate columns
train_data.drop(duplicate_col, axis=1, inplace=True)
test_data.drop(duplicate_col, axis=1, inplace=True)

# drop ID
train_data.drop('ID', axis=1, inplace=True)
test_id = test_data.ID
test_data.drop('ID', axis=1, inplace=True)

print "After data cleansing we have %d features in train data set" % (len(train_data.columns) - 1)
train_data.head()
```

## Feature Selection

```python
# inspired by ExtraTreesClassifier from sklearn http://scikit-learn.org/stable/modules/feature_selection.html
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.feature_selection import SelectFromModel

X = train_data.drop("TARGET",axis=1)
y = train_data.TARGET.values

clf = ExtraTreesClassifier(random_state=123)
clf = clf.fit(X, y)

# plot top 30 important features
feat_imp = pd.Series(clf.feature_importances_, index = X.columns.values).sort_values(ascending=False)
feat_imp[:30].plot(kind='bar', title='Feature Importances according to ExtraTreesClassifier', figsize=(12, 8))
plt.ylabel('Feature Importance Score')
plt.subplots_adjust(bottom=0.3)
plt.savefig('feature-importance.png')
plt.show()

model = SelectFromModel(clf, prefit=True)
X = model.transform(X)
test_data = model.transform(test_data)
print X.shape
print test_data.shape
```

## Model Selection

```python
# inspired by http://scikit-learn.org/stable/auto_examples/classification/plot_classifier_comparison.html
from sklearn import tree
from sklearn import metrics
from sklearn import ensemble
from sklearn import linear_model 
from sklearn import naive_bayes 
from sklearn import model_selection
from sklearn import neighbors
from sklearn import neural_network
from sklearn import svm
import time


models = {}

models['tree'] = tree.DecisionTreeClassifier()

models['extra_tree'] = ensemble.ExtraTreesClassifier()

models['forest'] = ensemble.RandomForestClassifier()

models['ada_boost'] = ensemble.AdaBoostClassifier()

models['bagging'] = ensemble.BaggingClassifier()

models['grad_boost'] = ensemble.GradientBoostingClassifier()

models['ridge'] = linear_model.RidgeClassifier()

models['passive'] = linear_model.PassiveAggressiveClassifier()

models['sgd'] = linear_model.SGDClassifier()

models['gaussian'] = naive_bayes.GaussianNB()

models['kneighbors'] = neighbors.KNeighborsClassifier()

models['mlp'] = neural_network.MLPClassifier()

##Too slow
#models['svc'] = svm.SVC()
models['linearSVC'] = svm.LinearSVC()

skf = model_selection.StratifiedKFold(shuffle=True, random_state=123)
score_metric = 'roc_auc'
scores = {}
compute_time = {}

def score_model(model):
    return model_selection.cross_val_score(model, X, y, cv=skf, scoring=score_metric)

for k, v in models.items():
    start_time = time.time()
    scores[k] = np.mean(score_model(v))
    compute_time[k] = time.time() - start_time

score_df = pd.DataFrame(scores.items(), columns=['Model', 'Score'], index=scores.keys())
score_df.plot(kind='bar', title='Model AUC Scores', figsize=(12, 8))
plt.ylabel('AUC Score')
plt.savefig('model-score.png')
plt.show()

compute_time_df = pd.DataFrame(compute_time.items(), columns=['Model', 'Time'], index=compute_time.keys())
compute_time_df.plot(kind='bar', title='Model Compute Time', figsize=(12, 8))
plt.ylabel('Seconds')
plt.savefig('model-compute-time.png')
plt.show()

print score_df
print compute_time_df
```

## Parameter tuning

```python
from sklearn.model_selection import GridSearchCV

gbc = ensemble.GradientBoostingClassifier(random_state=123)

# Explore learning_rate and n_estimators parameter
parameters = {'learning_rate':np.arange(0.01, 0.2, 0.05), 'n_estimators':np.arange(80, 200, 20)}
clf = GridSearchCV(gbc, parameters, cv=skf, scoring='roc_auc')
clf.fit(X, y)
print clf.best_params_
print clf.best_score_
```

## Submission Data

```python
probs = clf.predict_proba(test_data)

submission = pd.DataFrame({"ID":test_id, "TARGET": probs[:,1]})
submission.to_csv("submission.csv", index=False)

# ~0.8340 score on test data set
```
